## __object-oriented-style__

```python
x = np.linspace(0, 2, 100) 
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear') 
ax.plot(x, x**2, label='quadratic')  
ax.plot(x, x**3, label='cubic') 
ax.set_xlabel('x label')  #  x-label
ax.set_ylabel('y label')  # ylabel
ax.set_title("Simple Plot")  # axes title
ax.legend(); #legend

# layout='constrained' is throwing an error
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend();  # Add a legend.
```

# __Making a helper functions__

```python
def my_plotter(ax, data1, data2, param_dict):
    """
   function to make a graph
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)  #4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data3, data4, {'marker': 'x'})
my_plotter(ax2, data2, data1, {'marker': 'o'});
```

```python
data1, data2= np.random.randn(2, 100)  #2 random data sets
fig, (ax1) = plt.subplots(figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': '+'});
```

```python
# set_xlabel, set_ylabel, and set_title

mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7))
#histogram
n, bins, patches = ax.hist(x, 50, density=0.5, facecolor='C1', alpha=0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True);
```

## Working with multiple Figures and Axes

```python
fig, axd = plt.subplot_mosaic([['upleft', 'right'],
                               ['lowleft', 'right']],figsize=(10, 7) )
axd['upleft'].set_title('upleft')
axd['lowleft'].set_title('lowleft')
axd['right'].set_title('right');
```
